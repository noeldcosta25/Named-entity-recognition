```python
import spacy

# load english model
nlp = spacy.load("en_core_web_sm") # a pre-trained NLP model that already learned language patterns from huge text data.

text = input("Enter a sentence: ")

doc = nlp(text)

print("\nEntities Found:\n")

for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```
